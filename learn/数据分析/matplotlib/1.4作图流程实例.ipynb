```python
# 导入库与数据初始化
import numpy as np
import matplotlib.pyplot as plt

data = {'Barton LLC': 109438.50,
        'Frami, Hills and Schmidt': 103569.59,
        'Fritsch, Russel and Anderson': 112214.71,
        'Jerde-Hilpert': 112591.43,
        'Keeling LLC': 100934.30,
        'Koepp Ltd': 103660.54,
        'Kulas Inc': 137351.96,
        'Trantow-Barrows': 123381.38,
        'White-Trantow': 135841.99,
        'Will LLC': 104437.60}
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)
```

```python
# 粗劣作图
fig, ax = plt.subplots()   # 创建画布和区域
ax.barh(group_names, group_data)   # 应用bar函数，在区域（axes）中粗略作图
```

```python
# Matplotlib 中提供了许多绘图样式，以便您根据需要定制可视化效果。样式控制许多内容，例如颜色、线宽、背景等。要查看样式列表，我们可以使用style。
print(plt.style.available)
```

```python
# 改变样式
plt.style.use('fivethirtyeight')   # 设置绘图样式
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
```

```python
# 组件设置

# 首先，我们可以旋转x轴上的刻度标签。由于标签已经存在，因此要用 Axes.get_xticklabels() 去获取刻度标签对象本身。
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()

# 然后，我们可以运用 pyplot.setp() 一次性设置对象的多个属性。一般来说，get_ 和 setp 是配合使用的。
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels,                       # 被设置的对象
         rotation=45,                  # 旋转45度
         horizontalalignment='right')  # 水平方向，右对齐
```

```python
# 紧凑布局

plt.rcParams.update({'figure.autolayout': True})   # 紧凑图像边缘

fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
```

```python
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

ax.set(xlim=[-10000, 140000],    # 设置x轴范围
       xlabel='Total Revenue',   # 设置x轴标签
       ylabel='Company',         # 设置y轴标签
       title='Company Revenue')  # 设置标题

# 或逐个设置ax.set_
# ax.set_xlim([-10000, 140000])    # 设置x轴范围
# ax.set_xlabel('Total Revenue')   # 设置x轴标签
# ax.set_ylabel('Company')         # 设置y轴标签
# ax.set_title('Company Revenue')  # 设置标题

# 或运用plt.setp，这是最常用的函数和格式
# plt.setp(ax,
#          xlim=[-10000, 140000],    # 设置x轴范围
#          xlabel='Total Revenue',   # 设置x轴标签
#          ylabel='Company',         # 设置y轴标签
#          title='Company Revenue')  # 设置标题

# plt.setp的另一种书写格式，参照的是matlab
# plt.setp(ax,
#          'xlim', [-10000, 140000],    # 设置x轴范围
#          'xlabel', 'Total Revenue',   # 设置x轴标签
#          'ylabel', 'Company',         # 设置y轴标签
#          'title', 'Company Revenue')  # 设置标题

# 运用 get_ 和 setp 配合使用，不容易犯错。
```

```python
# 刻度标尺
def currency(x, pos):
    if x >= 1e6:
        s = '${:1.1f}M'.format(x*1e-6)
    else:
        s = '${:1.0f}K'.format(x*1e-3)
    return s

fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

ax.set(xlim=[-10000, 140000],    # 设置x轴范围
       xlabel='Total Revenue',   # 设置x轴标签
       ylabel='Company',         # 设置y轴标签
       title='Company Revenue')  # 设置标题
ax.xaxis.set_major_formatter(currency)   # 设置刻度便签的格式
plt.show()
```

```python
# 调整画布大小
fig, ax = plt.subplots(figsize=(8, 8))   # 设置画布大小
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
ax.xaxis.set_major_formatter(currency)
```

```python
# 组合多个效果
ax.axvline(group_mean, ls='--', color='r')   # 添加垂直竖线
for group in [3, 5, 8]:                      # 添加文本New Company
    ax.text(145000, group, "New Company", fontsize=10,
            verticalalignment="center")      # 垂直方向，中间对齐

ax.title.set(y=1.05)   # 调整标签位置
```

```python
# 所有代码
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    'figure.autolayout': True           # 紧凑显示
    })

plt.style.use('fivethirtyeight')    # 选择样式

data = {'Barton LLC': 109438.50,
        'Frami, Hills and Schmidt': 103569.59,
        'Fritsch, Russel and Anderson': 112214.71,
        'Jerde-Hilpert': 112591.43,
        'Keeling LLC': 100934.30,
        'Koepp Ltd': 103660.54,
        'Kulas Inc': 137351.96,
        'Trantow-Barrows': 123381.38,
        'White-Trantow': 135841.99,
        'Will LLC': 104437.60}
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)

fig, ax = plt.subplots(figsize=(8,8))   # 设置画布大小
ax.barh(group_names, group_data)

labels = ax.get_xticklabels()
plt.setp(labels,
         rotation=45,                   # 旋转45度
         horizontalalignment='right')   # 水平方向，右对齐

plt.setp(ax,
    xlim=[-10000, 140000],    # 设置x轴范围
    xlabel='Total Revenue',   # 设置x轴标签
    ylabel='Company',         # 设置y轴标签
    title='Company Revenue')  # 设置标题

def currency(x, pos):         # 自定义的“货币”函数
    if x >= 1e6:
        s = '${:1.1f}M'.format(x*1e-6)
    else:
        s = '${:1.0f}K'.format(x*1e-3)
    return s
ax.xaxis.set_major_formatter(currency)   # 设置刻度便签的格式

ax.axvline(group_mean, ls='--', color='r')   # 添加垂直竖线
for group in [3, 5, 8]:                      # 添加文本New Company
    ax.text(145000, group, "New Company", fontsize=10,
            verticalalignment="center")      # 垂直方向，中间对齐

ax.title.set(y=1.05)   # 调整标签位置

plt.show()
```
